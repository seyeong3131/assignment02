```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x=np.arange(0,12,0.01)

def function1(x) :
    return 2*x**2 + 3*x + 5

def derivation(f,x):
    h=1e-4
    return (f(x+h)-f(x-h))/2*h

x1=3
x2=7
x3=9
y=derivation(function1,x1)

def derivation2(x):
    return derivation(function1,x)

y1=(function1(x1))+(derivation(function1,x1)*(x-x1))+(derivation2(x1)/2*(x-x1)**2)
y2=(function1(x2))+(derivation(function1,x2)*(x-x2))+(derivation2(x2)/2*(x-x2)**2)
y3=(function1(x3))+(derivation(function1,x3)*(x-x3))+(derivation2(x3)/2*(x-x3)**2)

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,function1(x))
plt.show()
```
